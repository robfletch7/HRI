```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('form_data.csv')

df.columns
```

```python
df['Forename(s):'] = df['Forename(s):'].str.lower()
df['Surname:'] = df['Surname:'].str.lower()
df
```

```python
def make_label_df(df):
    
    label_df = pd.DataFrame()
    label_df['time'] = df['Start time']
    label_df['participantID'] = df.groupby(['Surname:']).ngroup()
    label_df['gender'] = df['How would you describe your gender?']
    before_emo = ['Stressed', 'Amused','Nervous', 'Engaged', 'Lost', 'Disappointed', 'Excited', 'Frustrated','Contempt']
    during_emo = ['Stressed2', 'Amused2','Nervous2', 'Engaged2', 'Lost2', 'Disappointed2', 'Excited2','Frustrated2', 'Contempt2']
    
    for i in range(len(before_emo)):
        before = 'before_' + before_emo[i].lower()
        during = 'during_' + before_emo[i].lower()
        change = 'change_in_'+ before_emo[i].lower()
        
        df.loc[df[before_emo[i]] == "Little to none", before] = 1
        df.loc[df[before_emo[i]] == "Moderate", before] = 2
        df.loc[df[before_emo[i]] == "High", before] = 3
        
        df.loc[df[during_emo[i]] == "Little to none", during] = 1
        df.loc[df[during_emo[i]] == "Moderate", during] = 2
        df.loc[df[during_emo[i]] == "High", during] = 3

        label_df[change] = np.sign(df[during]-df[before])
        
    return label_df
```

```python
labels = make_label_df(df)
```

```python
eda_data = pd.read_csv('average.csv')
```

```python
eda_data
```

```python
def make_eda_df(df):
    
    eda_df = pd.DataFrame()
    eda_df["eda_time"] = df["Participant Tags"]
    eda_df['perc_change'] = df['Percentage Change(before to during game)(%)']
    eda_df['abs_change'] = df['Average EDA During Game'] - df['Average EDA Before Game']
    
    return eda_df
```

```python
dat2 = eda_data.replace(r'^([A-Za-z]|[0-9]|_)+$', np.inf, regex=True)
```

```python
for col in eda_data.columns:
    dat2[col] = pd.to_numeric(eda_data[col], errors='coerce')
eda_2 = dat2.fillna(np.inf)
eda_2
```

```python
eda_feat = make_eda_df(eda_2)
eda_feat
```

```python
eda_feat = eda_feat.sort_values(by=['eda_time'])
eda_feat = eda_feat.reset_index(drop=True)
eda_feat
```

```python
labels = labels.sort_values(by=['time'])
labels = labels.reset_index(drop=True)
labels
```

```python
HR_df = pd.read_csv("Heart_rate_data_rob.csv")
HR_df.columns
```

```python
HR_feat = HR_df[["id",'perc_bpm', 'abs_bpm', 'perc_ibi', 'abs_ibi', 'perc_sdnn',
       'abs_sdnn', 'perc_sdsd', 'abs_sdsd', 'perc_rmssd', 'abs_rmssd',
       'perc_pnn20', 'abs_pnn20', 'perc_pnn50', 'abs_pnn50', 'perc_hr_mad',
       'abs_hr_mad', 'perc_sd1', 'abs_sd1', 'perc_sd2', 'abs_sd2', 'perc_s',
       'abs_s', 'perc_sd1/sd2', 'abs_sd1/sd2', 'perc_breathingrate',
       'abs_breathingrate']]
```

```python
HR_feat = HR_feat.sort_values(by=['id'])
HR_feat = HR_feat.reset_index(drop=True)
HR_feat = HR_feat.rename(columns={"id": "eda_time"})

```

```python
features = eda_feat.merge(HR_feat, on='eda_time',how='left')
features.replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
features
```

```python
features_labels = pd.DataFrame()
features_labels = pd.concat([features, labels],axis=1)
features_labels=features_labels.dropna().reset_index(drop=True)
features_labels.to_csv('final_dataset.csv')
```
